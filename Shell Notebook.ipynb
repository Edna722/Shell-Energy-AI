### Identify missing values

```python
# Function for missing values 
# A function to check for missing values in our dataset
def check_missing_values(data):
    # Count missing values in each column
    missing_values = data.isnull().sum()

    # Convert missing values count to percentage of total rows
    missing_percent = (missing_values / len(data)) * 100

    # Combine the missing values count and percent into a DataFrame
    missing_df = pd.concat([missing_values, missing_percent], axis=1)
    missing_df.columns = ['Missing Values', '% of Total']
# Return only columns with missing values
    missing_df = missing_df[missing_df['Missing Values'] > 0]

    return missing_df

# Check missing values in each dataset
display(check_missing_values(biomass_history))
display(check_missing_values(distance_matrix))
```

### Check for outliers 

```python
def handle_outliers_iqr(data, column, lower_bound_factor, upper_bound_factor, replace_with=None):
    """
    Handle outliers in a column using the Interquartile Range (IQR) method.

    Parameters:
        data (pd.DataFrame): DataFrame containing the data.
        column (str): Name of the column to handle outliers.
        lower_bound_factor (float): Lower bound factor to calculate the lower outlier threshold.
        upper_bound_factor (float): Upper bound factor to calculate the upper outlier threshold.
        replace_with: Value to replace outliers with (default is None, meaning outliers will be removed).

    Returns:
        pd.Series: Series with outliers handled.
    """
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    
    lower_bound = q1 - lower_bound_factor * iqr
    upper_bound = q3 + upper_bound_factor * iqr
    
    if replace_with is None:
        # Remove outliers
        data_filtered = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
        return data_filtered[column]
    else:
        # Replace outliers with specified value
        data[column] = data[column].apply(lambda x: replace_with if x < lower_bound or x > upper_bound else x)
        return data[column]

# Now you can use the preprocess_biomass_data function along with the handle_outliers_iqr function
biomass_history_cleaned = preprocess_biomass_data(biomass_history)

# Print the first few rows of the cleaned DataFrame
print(biomass_history_cleaned.head())
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Assuming you have a DataFrame named 'biomass_history' containing the biomass availability data
# with columns 'Index', 'Latitude', 'Longitude', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'

# Extract the relevant columns for scaling and normalization
biomass_features = biomass_history[['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']]

# Calculate the mean biomass availability for each harvesting site across all years
biomass_history['biomass_availability'] = biomass_features.mean(axis=1)

# Drop the individual year columns as they are no longer needed
biomass_history.drop(columns=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017'], inplace=True)
```

```python
# Perform Standard sclaing (Z-score normalization)
scaler = StandardScaler()
biomass_scaled = scaler.fit_transform(biomass_features)
```

```python
# Create a new DataFrame with the scaled biomass availability
biomass_scaled_df = pd.DataFrame(biomass_scaled, columns=['biomass_availability_scaled']*biomass_scaled.shape[1])

# Perform Min-Max Scaling (Normalization)
min_max_scaler = MinMaxScaler()
biomass_normalized = min_max_scaler.fit_transform(biomass_features)

# Create a new DataFrame with the normalized biomass availability
biomass_normalized_df = pd.DataFrame(biomass_normalized, columns=['biomass_availability_normalized']*biomass_normalized.shape[1])

# Concatenate the original DataFrame with the scaled and normalized DataFrames
biomass_history = pd.concat([biomass_history, biomass_scaled_df, biomass_normalized_df], axis=1)

# Create a new DataFrame with the normalized biomass availability
biomass_normalized_df = pd.DataFrame(biomass_normalized, columns=['biomass_availability_normalized']*biomass_normalized.shape[1])

# Concatenate the original DataFrame with the scaled and normalized DataFrames
biomass_history = pd.concat([biomass_history, biomass_scaled_df, biomass_normalized_df], axis=1)

# Now, let's create the new features for biomass forecasting
# Calculate the average biomass availability for each harvesting site over the years
average_biomass_per_site = biomass_history.groupby('Index')['biomass_availability'].mean()
biomass_history = biomass_history.merge(average_biomass_per_site, on='Index', suffixes=('', '_avg'))

# Assuming you want to create lag features for 1 year and 2 years
biomass_history['biomass_availability_lag1'] = biomass_history.groupby('Index')['biomass_availability'].shift(1)
biomass_history['biomass_availability_lag2'] = biomass_history.groupby('Index')['biomass_availability'].shift(2)

# Print the updated DataFrame
print(biomass_history.head())

```
